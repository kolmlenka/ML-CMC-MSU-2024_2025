<font color='LightSteelBlue'>**Подсказка**</font>

Значения <font color='CornflowerBlue'>**категориальной**</font> переменной определяют факт принадлежности к какой-то категории, категорий при этом должно быть ограниченное кол-во сильно меньшее размера датасета.

<font color='CornflowerBlue'>**Категориальная переменная**</font> - колонка, которая имеет ограниченное число уникальных значений. Вообще говоря, может быть как типа object (текстовые), так и int (численные). В случае int можно считать, что категории закодированы числами. В таком случае значения нельзя сравнивать между собой и совершать над ними арифметические операции. В этом задании категориальные переменные только типа object.

<font color='OrangeRed'>**Внимание!**</font> Для проверки корректности выведите кол-во пропусков во всех столбцах и общее кол-во строк в таблице - должна быть 3021 запись. Ошибка при удалении строк может привести к неправильному выполнению следующих заданий.




Возможно, пригодятся такие функции:
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html#pandas.DataFrame.fillna
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html#pandas.DataFrame.dropna

```python
# your code here
file_path = 'youtube_trending_mediterranean.csv'
data = pd.read_csv(file_path)
import pandas as pd

categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

data.fillna('', inplace=True)

data.dropna(subset=categorical_cols, inplace=True)

missing_counts = data.isnull().sum()
total_rows = len(data)

print("Количество пропусков в каждом столбце:\n", missing_counts)
print("Общее количество строк в таблице:", total_rows)
```

<font color='CornflowerBlue'>**Matplotlib**</font> - основная библиотека визуализации двумерных данных на языке python.

Рисунок создается с помощью последовательного вызова функций из модуля **pyplot** (создание рисунка, задание области рисования, построение графиков в области рисования, украшение графика метками). Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

    ✅ Все линии на графиках должны быть чётко видны

    ✅ Ограничивайте диапазоны значений там, где это возможно (например, вероятности можно ограничить диапазоном [0,1])
