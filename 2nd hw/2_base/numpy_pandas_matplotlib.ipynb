```python
# your code here
file_path = 'youtube_trending_mediterranean.csv'
data = pd.read_csv(file_path)
import pandas as pd

categorical_cols = data.select_dtypes(include=['object']).columns.tolist()

data.fillna('', inplace=True)

data.dropna(subset=categorical_cols, inplace=True)

missing_counts = data.isnull().sum()
total_rows = len(data)

print("    :\n", missing_counts)
print("    :", total_rows)
```

```python
import matplotlib.pyplot as plt #     

# your code here
unique_videos = data.drop_duplicates(subset=['video_id', 'trending_date'])
video_count_per_date = unique_videos.groupby('trending_date').size()

plt.figure(figsize=(3, 6))
video_count_per_date.plot(kind='line', marker='o')
plt.title('    ')
plt.xlabel('')
plt.ylabel(' ')
plt.xticks(rotation=0)
plt.grid()
plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
# your code here
unique_videos = data.drop_duplicates(subset='video_id')

video_count_by_country = unique_videos['country'].value_counts()

plt.figure(figsize=(12, 6))
video_count_by_country.plot(kind='bar', color='skyblue')
plt.title('    ')
plt.xlabel('')
plt.ylabel(' ')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
